```python
# Get all docs in a collection
docs = db.collection("wasteType").stream()

items_db = []
for doc in docs:
    items_db.append(doc.to_dict())
len(items_db)

```

```python
items_db[7]
```

```python
# Function to delete a collection
def delete_collection(coll_ref, batch_size):
    docs = coll_ref.limit(batch_size).stream()
    deleted = 0

    for doc in docs:
        print(f'Deleting doc {doc.id} => {doc.to_dict()}')
        doc.reference.delete()
        deleted += 1

    if deleted >= batch_size:
        return delete_collection(coll_ref, batch_size)

# Usage
collection_ref = db.collection('wasteType')
delete_collection(collection_ref, batch_size=10)

```

```python
# Reference to the collection
import json
collection_ref = db.collection('wasteType')

# Start a batch
batch = db.batch()

# Data to add
with open('cleaned_data.json', 'r') as file:
    data = json.load(file)

# Add documents to batch
for doc_data in data:
    doc_data['id'] = str(uuid.uuid4())
    doc_ref = collection_ref.document()  # Create a new document reference
    batch.set(doc_ref, doc_data)  # Add to batch

# Commit the batch
batch.commit()
```
